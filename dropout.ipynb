## CODE FOR FINDING VARIABLES WITH HIGH CORRELATION COEFFICIENTS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#pip install ucimlrepo
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
predict_students_dropout_and_academic_success = fetch_ucirepo(id=697) 
  
# data (as pandas dataframes) 
X = predict_students_dropout_and_academic_success.data.features 
y = predict_students_dropout_and_academic_success.data.targets 

# sns.pairplot(X)
# plt.title('Pair Plot of Features')
# plt.show()

# variable information 
#print(predict_students_dropout_and_academic_success.variables) 

correlation_matrix = X.corr()

column_names = X.columns
high_corr = 0
dict = {}

for i in range(len(column_names)):
    for j in range(i, len(column_names)):
        if correlation_matrix[column_names[i]][column_names[j]] > 0.7:
            if i != j:
                print(column_names[i], "and", column_names[j], "=", correlation_matrix[column_names[i]][column_names[j]])
                high_corr = high_corr + 1
                if column_names[i] not in dict:
                    dict[column_names[i]] = 1
                else:
                    dict[column_names[i]] = dict[column_names[i]] + 1
                
                if column_names[j] not in dict:
                    dict[column_names[j]] = 1
                else:
                    dict[column_names[j]] = dict[column_names[j]] + 1

print(high_corr)
print(dict)
```

# Generating some box plots for the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#pip install ucimlrepo
from ucimlrepo import fetch_ucirepo 

# fetch dataset 
predict_students_dropout_and_academic_success = fetch_ucirepo(id=697) 
  
# data (as pandas dataframes) 
X = predict_students_dropout_and_academic_success.data.features 
y = predict_students_dropout_and_academic_success.data.targets 

q1 = []
q2 = []
q3 = []
iqr = []
min = []
max = []

column_names = X.columns

for i, name in enumerate(column_names):
    q1.append(X[name].quantile(0.25))
    q2.append(X[name].quantile(0.5))
    q3.append(X[name].quantile(0.75))
    iqr.append(q3[i] - q1[i])
    min.append(q1[i] - 1.5 * iqr[i])
    max.append(q3[i] + 1.5 * iqr[i])

indices = set()

for i, feat in enumerate(column_names):
    for j in X[feat]:
        if j < min[i] or j > max[i]:
            indices.add(j)

print("Number of outliers =", len(indices))

plt.figure(figsize=(26, 46))
for i, name in enumerate(column_names, 1):
    plt.subplot(9, 4, i)
    plt.title(name)
    plt.boxplot(X[name])
plt.show()
```
