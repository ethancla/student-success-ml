## CODE FOR FINDING VARIABLES WITH HIGH CORRELATION COEFFICIENTS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#pip install ucimlrepo
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
predict_students_dropout_and_academic_success = fetch_ucirepo(id=697) 
  
# data (as pandas dataframes) 
X = predict_students_dropout_and_academic_success.data.features 
y = predict_students_dropout_and_academic_success.data.targets 

# Sets up the correlation Matrix
correlation_matrix = X.corr()

# Gets the column names and sets up the variables to keep track of the number of occurances
# each feature has and the number of highly correlated features
column_names = X.columns
high_corr = 0
dict = {}

# Nested for loops that check all correlations in the correlation matrix to see if they are
# above the 0.7 threshold we defined
for i in range(len(column_names)):
    for j in range(i, len(column_names)):
        if correlation_matrix[column_names[i]][column_names[j]] > 0.7:
            if i != j:
                print(column_names[i], "and", column_names[j], "=", correlation_matrix[column_names[i]][column_names[j]])
                high_corr = high_corr + 1
                if column_names[i] not in dict:
                    dict[column_names[i]] = 1
                else:
                    dict[column_names[i]] = dict[column_names[i]] + 1
                
                if column_names[j] not in dict:
                    dict[column_names[j]] = 1
                else:
                    dict[column_names[j]] = dict[column_names[j]] + 1

print(high_corr)
print(dict)
```

# Generating some box plots for the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#pip install ucimlrepo
from ucimlrepo import fetch_ucirepo 

# fetch dataset 
predict_students_dropout_and_academic_success = fetch_ucirepo(id=697) 
  
# data (as pandas dataframes) 
X = predict_students_dropout_and_academic_success.data.features 
y = predict_students_dropout_and_academic_success.data.targets 

# Sets up arrays to store values needed to create box plots for each feature
q1 = []
q2 = []
q3 = []
iqr = []
min = []
max = []

# Gets column names
column_names = X.columns

# holds name of non categorical features to find outliers of
names = ["Curricular units 2nd sem (credited)","Curricular units 2nd sem (enrolled)","Curricular units 2nd sem (evaluations)",
"Curricular units 2nd sem (approved)"]

# For loop that finds the values for each features box plot
for i, name in enumerate(names):
    q1.append(X[name].quantile(0.25))
    q2.append(X[name].quantile(0.5))
    q3.append(X[name].quantile(0.75))
    iqr.append(q3[i] - q1[i])
    min.append(q1[i] - 1.5 * iqr[i])
    max.append(q3[i] + 1.5 * iqr[i])

# sets up variables to keep track of the number of outliers found in each feature and
# a set of all the indices
indices = set()
times = 0

# for loop that finds the outliers based of IQR
for i, feat in enumerate(names):
    index = 0
    times = 0
    for j in X[feat]:
        if j < min[i] or j > max[i]:
            indices.add(index)
            times = times + 1
        index = index + 1
    print(feat, times)

# Prints the data of the outliers
print("Number of outliers =", len(indices))

print("Remaining Data poitns =", len(X['Marital Status']) - len(indices))

print("Outlier Indices:")
print(indices)

# Creates box plots of all the features
plt.figure(figsize=(26, 46))
for i, name in enumerate(column_names, 1):
    plt.subplot(9, 4, i)
    plt.title(name)
    plt.boxplot(X[name])
plt.show()
```
